```python
import os
import numpy as np

old_position = np.array([112.8,128.9])
new_position = np.array([74.6,184.5])
pos_fl = r'S:\12_04_2025_JenieSample\pos_1.txt'

import os
import numpy as np

new_pos_fl = os.path.splitext(pos_fl)[0]+'_modified.txt'
fid = open(new_pos_fl,'w')

for ln in open(pos_fl,'r'):
    if ',' in ln:
        x,y = np.array(eval(ln))+new_position-old_position
        fid.write(str(x)+','+str(y)+'\n')
fid.close()
print(new_pos_fl)

```
